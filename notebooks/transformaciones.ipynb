## Transformación de los datos

#### Eliminación de columnas con altos registros y eliminación de registros con pocas columnas nulas

```python
data = data.drop(columns=['ultima_actualizacion', 'fecha_inicio_liquidacion', 'fecha_fin_liquidacion', 'fecha_de_notificaci_n_de_prorrogaci_n'])
data = data.dropna()
data
```

#### Renombrar columnas

```python
data = data.rename(columns={'localizaci_n' : 'localizacion', 
                            'liquidaci_n' : 'liquidacion',
                            'obligaci_n_ambiental' : 'oblicacion_ambiental', 
                            'tipo_de_identificaci_n_representante_legal' : 'tipo_identificacion_representante_legal',
                            'identificaci_n_representante_legal' : 'identificacion_representante_legal',
                            'g_nero_representante_legal' : 'genero_representante_legal',
                            'sistema_general_de_regal_as' : 'sistema_general_de_regalias',
                            'recursos_propios_alcald_as_gobernaciones_y_resguardos_ind_genas_' : 'recursos_propios_alcaldias_gobernaciones_y_resguardos_indigenas', 
                            'duraci_n_del_contrato' : 'duracion_del_contrato',
                            'n_mero_de_cuenta' : 'numero_de_cuenta',
                            'n_mero_de_documento_ordenador_del_gasto' : 'numero_de_documento_ordenador_del_gasto',
                            'n_mero_de_documento_supervisor' : 'numero_de_documento_supervisor',
                            'n_mero_de_documento_ordenador_de_pago' : 'numero_de_documento_ordenador_de_pago'})
data
```

#### Formatos de fechas

```python
data["fecha_de_firma"] = pd.to_datetime(data["fecha_de_firma"], errors='coerce')
data["fecha_de_inicio_del_contrato"] = pd.to_datetime(data["fecha_de_inicio_del_contrato"], errors='coerce')
data["fecha_de_fin_del_contrato"] = pd.to_datetime(data["fecha_de_fin_del_contrato"], errors='coerce')
data
```

#### Formatos numéricos

```python
data["valor_del_contrato"] = pd.to_numeric(data["valor_del_contrato"], errors='coerce')
data["valor_de_pago_adelantado"] = pd.to_numeric(data["valor_de_pago_adelantado"], errors='coerce')
data["valor_facturado"] = pd.to_numeric(data["valor_facturado"], errors='coerce')
data["valor_pendiente_de_pago"] = pd.to_numeric(data["valor_pendiente_de_pago"], errors='coerce')
data["valor_pagado"] = pd.to_numeric(data["valor_pagado"], errors='coerce')
data["valor_amortizado"] = pd.to_numeric(data["valor_amortizado"], errors='coerce')
data["valor_pendiente_de"] = pd.to_numeric(data["valor_pendiente_de"], errors='coerce')
data["valor_pendiente_de_ejecucion"] = pd.to_numeric(data["valor_pendiente_de_ejecucion"], errors='coerce')
data["saldo_cdp"] = pd.to_numeric(data["saldo_cdp"], errors='coerce')
data["saldo_vigencia"] = pd.to_numeric(data["saldo_vigencia"], errors='coerce')
data["dias_adicionados"] = pd.to_numeric(data["dias_adicionados"], errors='coerce')
data["sistema_general_de_participaciones"] = pd.to_numeric(data["sistema_general_de_participaciones"], errors='coerce')
data["sistema_general_de_regalias"] = pd.to_numeric(data["sistema_general_de_regalias"], errors='coerce')
data["recursos_propios_alcaldias_gobernaciones_y_resguardos_indigenas"] = pd.to_numeric(data["recursos_propios_alcaldias_gobernaciones_y_resguardos_indigenas"], errors='coerce')
data["recursos_de_credito"] = pd.to_numeric(data["recursos_de_credito"], errors='coerce')
data["recursos_propios"] = pd.to_numeric(data["recursos_propios"], errors='coerce')

data
```

#### Eliminación de columnas innecesarias

```python
data = data.drop(columns=['localizacion'])
```

#### Tipos Booleanos

```python
data["es_grupo"].unique()
data["es_grupo"] = data["es_grupo"].astype(str).str.strip().str.lower().map({'si': True, 'no': False})

data["es_pyme"].unique()
data["es_pyme"] = data["es_pyme"].astype(str).str.strip().str.lower().map({'si': True, 'no': False})

data["liquidacion"].unique()
data["liquidacion"] = data["liquidacion"].astype(str).str.strip().str.lower().map({'si': True, 'no': False})

data["oblicacion_ambiental"].unique()
data["oblicacion_ambiental"] = data["oblicacion_ambiental"].astype(str).str.strip().str.lower().map({'si': True, 'no': False})

data["obligaciones_postconsumo"].unique()
data["obligaciones_postconsumo"] = data["obligaciones_postconsumo"].astype(str).str.strip().str.lower().map({'si': True, 'no': False})

data["reversion"].unique()
data["reversion"] = data["reversion"].astype(str).str.strip().str.lower().map({'si': True, 'no': False})

data["espostconflicto"].unique()
data["espostconflicto"] = data["espostconflicto"].astype(str).str.strip().str.lower().map({'si': True, 'no': False})

data["el_contrato_puede_ser_prorrogado"].unique()
data["el_contrato_puede_ser_prorrogado"] = data["el_contrato_puede_ser_prorrogado"].astype(str).str.strip().str.lower().map({'si': True, 'no': False})

data.info()
```

#### Mayúsculas

```python
data["nombre_entidad"] = data["nombre_entidad"].str.upper()
data["nombre_entidad"] = data["nombre_entidad"].str.replace("*", "", regex=False).str.replace("**", "", regex=False).str.replace("****", "", regex=False)

data["descripcion_del_proceso"] = data["descripcion_del_proceso"].str.upper()

data["proveedor_adjudicado"] = data["proveedor_adjudicado"].str.upper()

data["nombre_representante_legal"] = data["nombre_representante_legal"].str.upper()

data["objeto_del_contrato"] = data["objeto_del_contrato"].str.upper()

data["nombre_ordenador_del_gasto"] = data["nombre_ordenador_del_gasto"].str.upper()

data["nombre_supervisor"] = data["nombre_supervisor"].str.upper()

data["nombre_ordenador_de_pago"] = data["nombre_ordenador_de_pago"].str.upper()
```

#### URL

```python
data["urlproceso"] = data["urlproceso"].str.extract(r"url':\s*'([^']+)'")
data["urlproceso"]
```
